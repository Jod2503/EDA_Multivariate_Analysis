## Plots

there arrive 3 cases
1. Numerical-Numerical
2. Numerical- categorical
3. categorical-categorical

```python
# scatterplot
# for num-num, they should have same datatype
sns.scatterplot(x= tips['total_bill'],y= tips['tip'])
# the more the bill the increasing tips
```

```python
# if we also want for a particular gender
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
# gender for the individual
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'])
# seaborn does not support 3d plotting
```

```python
titanic.head()
```

```python
# avg age of people travelling in each Pclass
sns.barplot(x=titanic['Pclass'],y=titanic['Age'])
```

The marked black lines represent confidence intervals

CI- for repeated calculations of the same value, there is 95% chance the the value lies in this band.


```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'], hue=titanic['Sex'])
```

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'])
```

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
# male me younger survived more, female me older survived more
```

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'])
# age of people who didnt survive
sns.distplot(titanic[titanic['Survived'] == 1]['Age'])
# those who survived
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
sns.pairplot(titanic)
# combined scatterplot of all possible combinations
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```
